In the far field approximation, the array is positioned far enough from the singal source such that the signal forms and plane wave and subsequently the angle of incidence $\theta$ is equal between all array elements.
In this case, the path length difference between any two adjacent array elements is given by

$$\Delta l=d\sin\theta$$

Let $\tau$ be the time delay between the wave arriving at two adjacent array elements, given by

$$\tau = \frac{\Delta l}{c}$$

$$= \frac{d\sin\theta}{c}$$

The composite signal $y(t)$ between two adjacent array elements is given by

$$y(t) = s(t) + s(t-\tau)$$

The signal received at a given array element is described by

$$s(t) = m(t)\cos(\omega_0t)$$

where $m(t)$ is some function describing the modulation of the sinusoidal carrier wave with angular frequency $\omega_0$.

This gives us

$$y(t) = m(t)\cos(\omega_0 t) + m(t-\tau)\cos[\omega_0(t-\tau)]$$

Subsituting $\psi = \omega_0\tau$ and assuming that $\tau$ << $T_x$ where $T_x$ is the carrier period,

$$y(t) = m(t)\cos(\omega_0 t) + m(t)\cos(\omega_0t-\psi)$$

Note the approximation $m(t-\tau)\approx m(t)$ also comes from the assumption that the bandwidth of $m(t)$ << $\omega_0$.

Using phasor notation,

$$y(t) = m(t)+m(t)e^{-j\psi}$$

Recalling that $\omega = 2\pi f = \frac{2\pi c}{\lambda}$ and $\tau = \frac{d\sin\theta}{c}$, we have

$$\psi = 2\pi\frac{d}{\lambda}\sin\theta$$
where $\lambda$ = wavelength and $c$ = speed of light.

Thus the array output can be expressed as

$$y(t) = m(t)\sum_{i=1}^{2}e^{j(i-1)\psi}$$ 
$$= m(t)\sum_{i=1}^{2}e^{2\pi j(i-1)\frac{d}{\lambda}\sin\theta}$$

The summation term is referred to as the array factor $AF(\theta)$ where

$$AF(\theta) = \frac{y(t)}{m(t)}$$

Note that for an array of $n$ elements we get the geometric progression

$$= \sum_{i=1}^{n}e^{j(i-1)\psi}$$

$$= \frac{1-e^{jn\psi}}{1-e^{j\psi}}$$

$AF(\theta)$ is the ratio between the output and input of the array, i.e., it is the transfer function of the array.

By this token, it can be expressed as a power output

$$AF(\theta)_{dB} = 10\log\left(\left|{\frac{AF(\theta)}{2}}\right|^2\right)$$

Consider an array of uniformly spaced isotropic elements in the $x$-$y$ plane consisting of $M$ columns and $N$ rows. Let the spacing between consecutive elements in a row be $d_x$ and the element spacing in any given column be $d_y$.

We arbitrarily set the element at the origin to be element $(1,1)$, with the columns and rows increasing in the positive $x$ and $y$ directions respectively.

Consider a point $P$ located such that the distance from $P$ to $(p,q)$, $l_{p,q}$ is large enough so $l_{p,q}$ >> $Md_x$ and $l_{p,q}$ >> $Nd_y$ $\forall$ $(p,q)$. That is to say that $P$ is located in the far-field.

Recall that in the far-field approximation, EM waves originating at $P$ form plane waves that are incident on all array elements with the same angle of incidence $\theta$.

Let $t_{p,q}$ be the time taken for a signal to propagate from the emitter at $P$ to array element $(p,q)$.

Let $\hat{r}$ be the unit vector pointing from the origin to $P$ and let $\vec{r}_{pq}$ be the vector pointing from the origin to $(p,q)$

$$\hat{r} = r_x\hat{x} + r_y\hat{y} + r_z\hat{z}$$

$$\vec{r}_{pq} = (p-1)d_x\hat{x} + (q-1)d_y\hat{y}$$

In spherical coordinates, we have

$$r_x = \sin\theta \cos\phi$$

$$r_y = \sin\theta \sin\phi$$

$$r_z = \cos\theta$$

```python
# calculate projection of a onto b
def proj(a,b):
    p = (a.dot_product(b)/(norm(b)^2))*b
    return(p)

var('r_x', 'r_y', 'r_z', 'p', 'q', 'd_x', 'd_y')
r = vector([r_x, r_y, r_z])
r_pq = vector([(p-1)*d_x, (q-1)*d_y, 0])
# s = proj(r_pq, r)
r.dot_product(r_pq)
```

The path length difference between $l_{1,1}$ and $l_{p,q}$ can be found by calculating the projection of $r_{pq}$ onto $\hat{r}$. Projection of a vector onto a unit vector is simply the dot product.

$$\Delta l_{pq} = d_{11} - d_{pq}$$ 

$$= proj_{\hat{r}}\vec{r}_{pq} = \frac{\hat{r} \cdot \vec{r}_{pq}}{\lVert{\hat{r}\rVert^2}} = \hat{r} \cdot \vec{r}_{pq}$$

$$\Delta l_{pq}= d_x(p-1)\sin\theta \cos\phi + d_y(q-1)\sin\theta \sin\phi$$

Note also that the time difference between paths is given by

$$\tau = \frac{\Delta l_{pq}}{c}$$

Let $s_{11}(t)$ be the signal received at element $(1,1)$.

$$s_{11}(t) = m(t)\cos(\omega_0 t)$$

If the signal is received at $(p,q)$ earlier by some time $\tau_{pq}$ then we can say

$$s_{pq}(t) = s_{11}(t + \tau_{pq})$$

$$= m(t + \tau_{pq})\cos[\omega_0(t + \tau_{pq})]$$

$$\cong m(t)e^{j\omega_0t}e^{j\omega_0\tau_{pq}}$$

In phasor notation,

$$s_{pq}(t) = m(t)e^{j\omega_0\tau_{pq}}$$

Again, we assume that $m(t)$ evolves much slower than the carrier wave. It is also implied that we are considering $Re[s_{pq}(t)]$.

Recall that $\omega_0 = 2\pi \frac{c}{\lambda}$ and $\tau_{pq} = \frac{\Delta l_{pq}}{c}$ we have

$$s_{pq}(t) = m(t)e^{j2\pi \frac{\Delta l_{pq}}{\lambda}}$$

Substituing $\Delta l_{pq}= d_x(p-1)\sin\theta \cos\phi + d_y(q-1)\sin\theta \sin\phi$,

$$s_{pq}(t) = m(t)e^{j\frac{2\pi}{\lambda} [d_x(p-1)\sin\theta \cos\phi + d_y(q-1)\sin\theta \sin\phi]}$$

$$=m(t)e^{j\frac{2\pi}{\lambda} d_x(p-1)\sin\theta \cos\phi}e^{jd_y(q-1)\sin\theta \sin\phi}$$
